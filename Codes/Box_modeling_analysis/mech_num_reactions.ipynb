```python
import matplotlib.pyplot as plt

# Data to be plotted
models = ['SAPRC-18', 'MOZART-T1', 'GEOS-Chem', 'MCM']
reactions_counts = [1800, 350, 800, 17000]

# Sorting data based on reactions count in descending order
sorted_data = sorted(zip(reactions_counts, models), reverse=True)
sorted_reactions_counts, sorted_models = zip(*sorted_data)

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_models, sorted_reactions_counts, color='black')

# Adding titles and labels
plt.title('Number of Reactions in Different Chemical Mechanisms', fontsize=16)
plt.xlabel('Chemical Mechanism', fontsize=14)
plt.ylabel('Number of Reactions', fontsize=14)
plt.yscale('log')  # Log scale to better display the large difference

# Annotating each bar with the number of reactions
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval * 1.04, f'{int(yval)}', 
             va='bottom', ha='center', color='black', fontweight='bold')

# Show plot
plt.show()
```
