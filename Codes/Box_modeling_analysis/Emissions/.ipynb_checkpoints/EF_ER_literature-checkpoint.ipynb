```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# =====================================
# Andreae 2019, copy from the raw paper
# =====================================
# read the data
# be careful about the totol NMOC lines
df = pd.read_csv('Andreae2019_EF.csv', sep='\t', index_col = 'Species')
# reduced the species 
studied_spec= ['CO', 'Furan', '2-Methyl-furan', '3-Methyl-furan', 
               '2-Ethylfuran', '2,4-Dimethyl-furan', '2,5-Dimethyl-furan', 
               'Furfural (2-Furaldehyde)']
df = df.loc[studied_spec]

# reduced columns
dropped_fuel = ['Species.1',
                'Dung', 'std.dev..7', 'N.7', 
                'Charcoal making', 'std.dev..8', 'N.8', 
                'Charcoal burning', 'std.dev..9', 'N.9', 
                'Garbage burning', 'std.dev..10', 'N.10', 
                'Lab studies', 'std.dev..11', 'N.11']
df = df.drop(dropped_fuel, axis=1)

# ignore isomers for 2,5-dimethylfuran and 2-methylfuran for now
#df = df.drop(['3-Methyl-furan', '2-Ethylfuran', '2,4-Dimethyl-furan'])
#df = df.drop(['2-Ethylfuran', '2,4-Dimethyl-furan'])

# playing with missing values and range of standard deviation
for ind in df.index:
    for col in df.columns:
        # deal with methods
        if col == 'EMb' and '---' in df.loc[ind, col]: df.loc[ind, col] = 'Lit.'
        # deal with nan values
        if ' - ' in str(df.loc[ind, col]): df.loc[ind, col] = 0
        # deal with range of std
        if '-' in str(df.loc[ind, col]):
            data1, data2 = str(df.loc[ind, col]).split('-')
            avg_std = np.sqrt((float(data1)**2 + float(data2)**2)/np.sqrt(2))
            df.loc[ind, col] = avg_std
            
# playing with lumping speices
for col in df.columns:
    if 'std.dev.' in col:
        # 2-methylfuran and 3-methylfruan
        data_list = [float(df.loc['2-Methyl-furan', col]), float(df.loc['3-Methyl-furan', col])]
        numpt     = len(data_list)-sum(pd.array(data_list).isna())
        df.loc['Methylfurans', col] = np.sqrt(sum([data**2 for data in data_list]))/np.sqrt(numpt)

        # C2 furans
        data_list = [float(df.loc['2,4-Dimethyl-furan', col]), float(df.loc['2,5-Dimethyl-furan', col]), float(df.loc['2-Ethylfuran', col])]
        numpt     = len(data_list)-sum(pd.array(data_list).isna())
        df.loc['C2 furans', col] = np.sqrt(sum([data**2 for data in data_list]))/np.sqrt(numpt)
        
        
    elif 'N' in col or 'EMb' in col:
        df.loc['Methylfurans', col] = np.nan
        df.loc['C2 furans', col] = np.nan
    else:
        df.loc['Methylfurans', col] = float(df.loc['2-Methyl-furan', col])  + float(df.loc['3-Methyl-furan', col])
        df.loc['C2 furans', col] = float(df.loc['2,4-Dimethyl-furan', col])  + float(df.loc['2,5-Dimethyl-furan', col]) + float(df.loc['2-Ethylfuran', col])
df_andreae = df
df_andreae = df_andreae.rename(index={'Furfural (2-Furaldehyde)':'Furfurals'})
```

```python
# =========================================================================
# Koss 2018
# copy from the raw data
# mapping Lodgepole pine to boreal forest, this is suggested in Hayden2022
# mapping Bear Grass to Savanna
# mapping Rice Straw to Agriculture
# mapping peat to peanland
# =========================================================================
# read data
df = pd.read_csv('Koss2018_EF.csv', on_bad_lines='skip', sep='\t', index_col = 'NMOG identity')

# reduced cols
studied_cols = ['Lodgepole pine', 'Standard deviation.1', 
                'Bear Grass', 'Standard deviation.14', 
                'Rice Straw', 'Standard deviation.15',
                'Peat', 'Standard deviation.16']
df = df[studied_cols]

# reduced spec
studied_spec= ['Furan', '2-methylfuran + 3-methylfuran + general HCO', '2,5-dimethyl furan + 2-ethylfuran + other C2 substituted furans',
               '2-furfural + 3-furfural + other HCO2',
               '5-Methyl furfural +Benzene diols (=catechol, resorcinol)']
df = df.loc[studied_spec, :]
df_koss = df

# ============================
# Permar 2021 temperate forest
# copy from the raw data
# ============================
df = pd.read_csv('Permar2021_EF.csv', on_bad_lines='skip', sep='\t', index_col = 'VOC contributorc')
# reduced columns
dropped_cols = ['Dominant isotopologue exact mass (da)b', 'Chemical formula']
df = df.drop(dropped_cols, axis = 1)

# reduced species
studied_spec = ['Carbon monoxide', 'Furan', 
                '2-Methylfuran, 3-Methylfuran', '2-Methylfurand', '3-Methylfurand',
                'C2-Substituted furan isomers, 2,5-Dimethylfuran, 2-Ethylfuran',
                '2-Furfural (=furaldehyde), 3-Furfural', '5-Methylfurfural, Benzene diols (=catechol, resorcinol)']
df = df.loc[studied_spec]

# playing with lumping speices
for col in df.columns:
    if col == 'Standard deviation':
        df.loc['Methylfurans_TOGA', col] = np.sqrt(float(df.loc['2-Methylfurand', col])**2 + float(df.loc['3-Methylfurand', col])**2)/np.sqrt(2) 
    else:
        df.loc['Methylfurans_TOGA', col] = float(df.loc['2-Methylfurand', col])  + float(df.loc['3-Methylfurand', col])
df_permar = df


# trim the compounds to be consistent with Andreae for Permar and Hayden dataset
# methylfurans from TOGA is saved for later usage
Methylfurans_TOGA_permar = df_permar.loc['Methylfurans_TOGA', :]
df_permar = df_permar.drop(['2-Methylfurand', '3-Methylfurand', 'Methylfurans_TOGA'])

# =============================================================================
# Hayden 2022 boreal forest
# This paper does not have a excel supplement so that we mannually get the data
# =============================================================================
# Let's say we want data for CO, Furan, 
# 2-Methylfuran, 3-Methylfuran, 
# C2-Substituted furan isomers, 2,5-Dimethylfuran,
# 2-Furfural, 3-Furfural, 5-Methylfurfural
names = ['Carbon monoxide', 'Furan', 
         '2-Methylfuran, 3-Methylfuran', 
         'C2-Substituted furan isomers, 2,5-Dimethylfuran, 2-Ethylfuran', 
         'Furfurals', '5-Methylfurfural, Benzene diols (=catechol, resorcinol)']
avg_EF= [116, 0.39, 
         0.3, 
         0.087, 
         0.65, 0.11]
std_EF= [6, 0.19, 
         0.15, 
         0.044, 
         0.31, 0.005]
data = {'Emission factor': avg_EF,
        'Standard deviation': std_EF}
df = pd.DataFrame(data)
df.index = names
df_hayden = df

# ============================================
# Further reduce variables for these datasets
# ============================================
# Ignore CO for now as the different y scale
df_andreae = df_andreae.drop('CO', axis=0)
df_permar  = df_permar.drop('Carbon monoxide', axis = 0)
df_hayden  = df_hayden.drop('Carbon monoxide', axis = 0)

# Also ignore methylfuran and C2 furan isomers as they are just lumped
df_andreae = df_andreae.drop(['2-Methyl-furan', '3-Methyl-furan', '2,4-Dimethyl-furan', '2,5-Dimethyl-furan', '2-Ethylfuran'], axis=0)
```

```python
# =====================
# get data from andreae
# =====================
col_andreae_EF  = [col for col in df_andreae.columns if ('std.dev.' not in col) & ('N' not in col) & ('EMb' not in col)]
col_andreae_std = [col for col in df_andreae.columns if ('std.dev.' in col)] 
col_andreae_num = [col for col in df_andreae.columns if ('N' in col)] 
df_andreae_EF   = df_andreae[col_andreae_EF].astype('float')
df_andreae_std  = df_andreae[col_andreae_std].astype('float')
df_andreae_num  = df_andreae[col_andreae_num].astype('float')
# rename the columns to minimize the space
df_andreae_EF.columns = ['Savanna', 'Tropical', 'Temperate', 'Boreal', 'Peatland', 'Cropland', 'Biofuel']
df_andreae_std.columns = ['Savanna', 'Tropical', 'Temperate', 'Boreal', 'Peatland', 'Cropland', 'Biofuel']
df_andreae_num.columns = ['Savanna', 'Tropical', 'Temperate', 'Boreal', 'Peatland', 'Cropland', 'Biofuel']

# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# adding methylfurfural based on fuel distribution
# this process may be not right
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#for col in df_andreae_EF.columns:
#    fuel_ratio = df_andreae_EF.loc['Furfurals', col]/df_andreae_EF.loc['Furfurals', 'Temperate']
#    df_andreae_EF.loc['5-Methylfurfural, Benzene diols', col] = df_permar.loc['5-Methylfurfural, Benzene diols (=catechol, resorcinol)', 'Emission factor'] * fuel_ratio

df_andreae_EF.loc['5-Methylfurfural, Benzene diols', :] = [np.nan]*len(df_andreae_EF.columns)
df_andreae_std.loc['5-Methylfurfural, Benzene diols', :] = [np.nan]*len(df_andreae_std.columns)
df_andreae_num.loc['5-Methylfurfural, Benzene diols', :] = [np.nan]*len(df_andreae_num.columns)

# =====================
# get data from Permar
# =====================
# modify permar's compound names to be consistent with Andreae
SpecMapping2Andreae_1 = {'2-Methylfuran, 3-Methylfuran': 'Methylfurans',
                         'C2-Substituted furan isomers, 2,5-Dimethylfuran, 2-Ethylfuran': 'C2 furans',
                         '2-Furfural (=furaldehyde), 3-Furfural': 'Furfurals', 
                         '5-Methylfurfural, Benzene diols (=catechol, resorcinol)': '5-Methylfurfural, Benzene diols'}
df_hayden = df_hayden.rename(index=SpecMapping2Andreae_1)
df_permar = df_permar.rename(index=SpecMapping2Andreae_1)




# make the data consistent with andreae
SpecMapping2Andreae_2 = {'2-methylfuran + 3-methylfuran + general HCO': 'Methylfurans',
                         '2,5-dimethyl furan + 2-ethylfuran + other C2 substituted furans': 'C2 furans',
                         '2-furfural + 3-furfural + other HCO2': 'Furfurals', 
                         '5-Methyl furfural +Benzene diols (=catechol, resorcinol)': '5-Methylfurfural, Benzene diols'}
df_koss = df_koss.rename(index=SpecMapping2Andreae_2)
```

```python
# =================
# Liu et al., 2017
# =================
df_liu_EF = pd.DataFrame(columns=df_andreae_EF.columns, index = df_andreae_EF.index)
for ind in df_liu_EF.index:
    for col in df_liu_EF.columns:
        if col == 'Temperate': 
            df_liu_EF.loc[ind, col] = 0.51
df_liu_std = pd.DataFrame(columns=df_andreae_std.columns, index = df_andreae_std.index)
for ind in df_liu_std.index:
    for col in df_liu_std.columns:
        if col == 'Temperate': 
            df_liu_std.loc[ind, col] = 0.06
# =================
# Koss et al., 2018
# =================
df_koss_EF = pd.DataFrame(columns=df_andreae_EF.columns, index = df_andreae_EF.index)
for ind in df_koss_EF.index:
    for col in df_koss_EF.columns:
        if col == 'Boreal': 
            df_koss_EF.loc[ind, col] = df_koss.loc[ind, 'Lodgepole pine']
        if col == 'Savanna':
            df_koss_EF.loc[ind, col] = df_koss.loc[ind, 'Bear Grass']
        if col == 'Cropland':
            df_koss_EF.loc[ind, col] = df_koss.loc[ind, 'Rice Straw']
        if col == 'Peatland':
            df_koss_EF.loc[ind, col] = df_koss.loc[ind, 'Peat']


df_koss_std = pd.DataFrame(columns=df_andreae_std.columns, index = df_andreae_std.index)
for ind in df_koss_std.index:
    for col in df_koss_std.columns:
        if col == 'Boreal': 
            df_koss_std.loc[ind, col] = df_koss.loc[ind, 'Standard deviation.1']
        if col == 'Savanna':
            df_koss_std.loc[ind, col] = df_koss.loc[ind, 'Standard deviation.14']
        if col == 'Cropland':
            df_koss_std.loc[ind, col] = df_koss.loc[ind, 'Standard deviation.15']
        if col == 'Peatland':
            df_koss_std.loc[ind, col] = df_koss.loc[ind, 'Standard deviation.16']
            
# ===================
# permar et al., 2021
# ===================
df_permar_EF = pd.DataFrame(columns=df_andreae_EF.columns, index = df_andreae_EF.index)
for ind in df_permar_EF.index:
    for col in df_permar_EF.columns:
        if col == 'Temperate': 
            df_permar_EF.loc[ind, col] = df_permar.loc[ind, 'Emission factor']

df_permar_std = pd.DataFrame(columns=df_andreae_std.columns, index = df_andreae_std.index)
for ind in df_permar_std.index:
    for col in df_permar_std.columns:
        if col == 'Temperate': 
            df_permar_std.loc[ind, col] = df_permar.loc[ind, 'Standard deviation']
            
# ===================
# Hayden et al., 2022
# ===================
df_hayden_EF = pd.DataFrame(columns=df_andreae_EF.columns, index = df_andreae_EF.index)
for ind in df_hayden_EF.index:
    for col in df_hayden_EF.columns:
        if col == 'Boreal': 
            df_hayden_EF.loc[ind, col] = df_hayden.loc[ind, 'Emission factor']

df_hayden_std = pd.DataFrame(columns=df_andreae_std.columns, index = df_andreae_std.index)
for ind in df_hayden_std.index:
    for col in df_hayden_std.columns:
        if col == 'Boreal': 
            df_hayden_std.loc[ind, col] = df_hayden.loc[ind, 'Standard deviation']
```

```python
# next step: plot the error bar plot for these 5 compounds
# plotting setting
fontsize_title = 28
fontsize_label = 30
fontsize_legend = 15
fontsize_tick = 15
fontsize_subtitle = 20
fotnsize_numpt = 12
error_bar_capsize = 6
n_rows, n_cols = 2, 3
linewidth = 3

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16,12),sharex=True, sharey=False)

col_num = np.arange(len(df_andreae_EF.columns)) 

for ind, ax in zip(df_andreae_EF.index, axes.flatten()):
    
    ax.errorbar(col_num-0.2, df_liu_EF.loc[ind, :], yerr = df_liu_std.loc[ind, :], 
                fmt = 'o', color = 'C0', alpha = 1,
                label = 'Liu et al. (2017)',
                capsize = error_bar_capsize)
    
    ax.errorbar(col_num-0.1, df_koss_EF.loc[ind, :], yerr = df_koss_std.loc[ind, :], 
                fmt = 'o', color = 'C1', alpha = 1,
                label = 'Koss et al. (2018)',
                capsize = error_bar_capsize)
    
    ax.errorbar(col_num, df_andreae_EF.loc[ind, :], yerr = df_andreae_std.loc[ind, :], 
                fmt = 'o', color = 'black', alpha = 1,
                label = 'Andreae (2019)',
                capsize = error_bar_capsize)
    
    ax.errorbar(col_num+0.1, df_permar_EF.loc[ind, :], yerr = df_permar_std.loc[ind, :], 
                fmt = 'o', color = 'C2', alpha = 1,
                label = 'Permal et al. (2021)',
                capsize = error_bar_capsize)
    
    ax.errorbar(col_num+0.2, df_hayden_EF.loc[ind, :], yerr = df_hayden_std.loc[ind, :], 
                fmt = 'o', color = 'C3', alpha = 1,
                label = 'Hayden et al. (2022)',
                capsize = error_bar_capsize)
    
    # title
    ax.set_title(ind, fontsize=fontsize_subtitle)
    # x and y tick
    ax.tick_params(axis='both', labelsize=fontsize_tick)
    
    ax.set_xticks(np.arange(len(df_andreae_EF.columns)))
    ax.set_xticklabels(df_andreae_EF.columns, rotation = 30) 
    # legend size
    if ind == 'Furan':ax.legend(fontsize=11)
    
# remove the useless ax
fig.delaxes(axes[1][2])


# set labels
fig.text(0.05, 0.5, "Emission factor (g kg ${^{-1}}$ DM)", 
         va="center",fontsize=25, 
         color ='black',rotation=90)
fig.text(0.5, 0.03, "Fuel type", 
         ha="center",fontsize=25, 
         color ='black')

fig.text(0.2,0.92, 'Biomass burning emission factor of furanic compounds', fontsize=fontsize_title)


#fig.text(0.8, 0.8, "Emission ratio", ha="center", va="center", fontsize = 20, color ='red')
#plt.xlabel("Emission estimates (Gg)", size=25)
#plt.ylabel("", size=25)
plt.show()
```
