```python
# compounds we would look into
#compounds = ['Lumped_C>=3_aldehydes', 'MEK']
# compounds we would look into
#Flight_IDs   = ['P-3B', 'RF03', 'FN19']
Flight_IDs   = ['P-3B', 'RF03', 'RF07', 'RF09', 'FN19']

compounds = ['O3', 'Ox']
#compounds = ['HNO4', 'HNO3']
#compounds = ['NOx', 'NOx']
compounds = ['HCOOH', 'MEK']
compounds = ['CH2O', 'GLYC', 'O3']
compounds = ['Ox', 'Ox']

#compounds = ['HO2', 'RO2']

is_prod = 0
topX=15

# Choice of mechanism (Editable)
mech = 'MCM_GCvocs'
#mech = 'GC_base'
mech = 'MCM_base_FUR'

# plotting for the figure
linewidth = 3
error_infor = False
fontsize_flight = 20
fontsize_tick   = 20
fontsize_lable  = 10
fontsize_comp   = 20
fontsize_pct    = 16
# subplots 
fig, axes = plt.subplots(len(compounds), len(Flight_IDs), 
                         figsize=(7*len(Flight_IDs), 4*len(compounds)),
                         sharex='col')
for row, compound in enumerate(compounds):
    for col, Flight_ID in enumerate(Flight_IDs):
        # ------------
        # Reading data
        # ------------
        # model files
        # for simple compounds
        if mech == 'MCM_base_FUR': 
            df_prod = pd.read_csv(f'{file_prefix}{compound}/{Flight_ID}_Production_rates_full_MCMBBVOC.csv', index_col=0)
            df_loss = pd.read_csv(f'{file_prefix}{compound}/{Flight_ID}_Loss_rates_full_MCMBBVOC.csv', index_col=0)
            
        # we need to further preprocess the data if we want to look into O3.
        if compound in ['O3', 'Ox']:
            # group model data
            data_prod = O3_PL_budget_processor(df_prod)
            data_loss = -O3_PL_budget_processor(df_loss)
        else:
            # group model data
            data_prod = df_prod
            data_loss = -df_loss
        
        # Filter the data to find elements that follow the pattern X + O3
        #filtered_data = [item for item in data_loss.columns if "O + NO2" in item]
        #print(filtered_data)

        # define the base colormap
        colors = plt.cm.Paired(np.linspace(0, 1, topX+1))
        cmap = ListedColormap(colors)  
        
        data_plot = data_prod if is_prod else data_loss 
        
        # map the term
        data_plot = prd_map_process(data_plot)


        # make top 10 and others
        tmp = get_mean_topX_df(data_plot, 100, ascending=False).mean() 
        
        # Specified reactions for the first tmp sum
        specified_reactions = ["O1D", "HO2 + O3", "HNO3", "OH + O3", "NO3 + hv", "O + NO2", "HO2 + NO3"]
        key_VOCs  = ["Furanoid", "Monoterpenes", "BCARY + O3", "Catechol", "CATECHOL + NO3", "OXCATEC1O + O3", "MCATEC1O + O3"]
        # Calculating the two tmp sums
        tmp_sum_specified = sum(value for reaction, value in tmp.items() if reaction in specified_reactions)
        tmp_sum_remaining = sum(value for reaction, value in tmp.items() if reaction not in specified_reactions)
        tmp_sum_keyVOCs   = sum(value for reaction, value in tmp.items() if reaction in key_VOCs)
        print(100*tmp_sum_remaining/(tmp_sum_remaining+tmp_sum_specified), 100*tmp_sum_keyVOCs/(tmp_sum_remaining))
        
        data_plot = get_mean_topX_df(data_plot, topX, ascending=False).mean()  

        data_pct  = 100*data_plot/data_plot.sum()
        
        
        # -----------------
        # Plotting figures
        # -----------------
        # Bar plot, the unit here is ppb/h
        (data_plot).plot.barh(ax = axes[row, col], stacked=True, 
                                   color=colors,
                                   #colormap= cmap,
                                   edgecolor='black', linestyle="-", width = 0.7)
        
        '''
        # Pie chart
        # Create a pie chart within the stacked bar plot
        radius = 0.9
        pie_ax = axes[row, col].inset_axes([0.2, 0.05, radius, radius])  # x, y, width, height
        
        # Create labels for the pie chart with a threshold of 5% for displaying text
        labels = []
        for comp, percent in zip(data_plot.keys().values, data_pct):
            if percent >= 5: # comment of the percentage
                labels.append(f"{data_plot[comp]:.1f} ({percent:.0f}%)")
            else:
                labels.append(None)  # Set label to None for components with < 5%

        # Create the pie chart
        pie, _ = pie_ax.pie(data_plot, autopct=None, startangle=90, labels=None)
        # Calculate label positions and rotation within each slice
        radius = 0.5
        # Calculate label positions within each slice
        for wedge, label_text in zip(pie, labels):
            angle = (wedge.theta1 + wedge.theta2) / 2.0
            x = radius * np.cos(np.deg2rad(angle))
            y = radius * np.sin(np.deg2rad(angle))

            # Adjust the rotation angle for the label
            rotation_angle = np.degrees(np.arctan2(y, x))
            # Correct the angle so that text is always right-side-up
            if angle > 90 and angle < 270:
                rotation_angle -= 180
            # Set the label's position
            pie_ax.annotate(label_text, (x, y), ha='center', va='center', fontsize=12, rotation=rotation_angle)
        # Customize the pie chart appearance as needed
        pie_ax.set_aspect('equal')  # Set aspect ratio to make it a circle
        '''
        # set up the title
        if row == 0: axes[row, col].set_title(id2fire_name[Flight_ID], loc ='center', size=fontsize_flight)
            
        # Set compound name as y-axis label for the first column of each row
        if col == 0:
            axes[row, col].set_ylabel(text_ysublabels[compound], fontsize=fontsize_comp*1.2, rotation=90, labelpad=15)
            # Align y-axis labels at the same x location
            axes[row, col].yaxis.set_label_coords(-0.6, 0.5)
            
        # Spacing Out
        plt.subplots_adjust(wspace=.20, hspace=.10)
        axes[row, col].tick_params(axis='both', labelsize=fontsize_tick)
                

# tight the layout
fig.tight_layout()

# Texts
if is_prod: 
    fig.text(0.45,-0.01, 'Production rate (ppb h$^{-1})$', ha="center", va="center", rotation=0, fontsize=30)
else:
    fig.text(0.45,-0.01, 'Loss rate (ppb h$^{-1})$', ha="center", va="center", rotation=0, fontsize=30)

# Specify the file path and name for the saved figure
output_file_path = '/glade/work/lixujin/PYTHON/SciProj/Box_modeling_analysis/PL_budget/figures/PL.jpg'  # Update the path and file name

# Save the figure
fig.savefig(output_file_path, format='jpg', bbox_inches='tight')

# Optional: Show the plot if you want to view it in a notebook or a Python script
plt.show()
```
