```python
################################################################################
# Basic settings including compounds, flight ID, running mehcanism and others
################################################################################
obs_only=0
# 1) species list (customized) 
# addding other compounds such as phenol
# For HOx, TRO2, TPNs, TANs, TotherNOy, TVOCR, TNOy, and TNOz, the have been included in the processing script. Same for OH and HO2
spec_total = [
    'Ethyne', 
    'Ethene', 'Lumped C>=3 alkenes',
    'Ethane', 'Propane', 'Lumped C>=4 alkanes', 
    'Formaldehyde', 'Acetaldehyde', 'Lumped C>=3 aldehydes',
    'Glyoxal', 'Methylglyoxal', 'Benzene', 'Toluene', 'Xylenes',
    'Formic acid',  'Acetic acid',
    'Isoprene', 'Monoterpenes', 'Methacrolein', 'MVK',
    'Methanol', 'Ethanol',
    'Acetone', 'MEK', 'Glycoaldehyde',
    'DMS', 'Hydroxyacetone', 
    'Phenol', 'Cresol', 'Benzaldehyde', 
    'Furans', 'Furfurals', # other VOCs,
    'Furan', 'Methylfuran', 'Dimethylfuran', 'Furfural', 'Methylfurfural',
    'Furanone', 'Maleic anhydride', 
    'Acrolein', 'Butadiene', 'Butenal', 'Butanedione', 
    'Pentan2one', 'Pentan3one', 'Catechol', 'Hexanones', 'Guaiacol', 'Heptanone',
    'Styrene', 'Tolualdehyde', 'Creosol', 'Syringol', 
    'Trimethyl123benzene', 'Isopropylbenzene', 'Propylbenzene', 'Methyltoluene', 
    'Pethyltoluene', 'Trimethyl135benzene', 'Oethyltoluene', 
    'Pinic acid', 'Pinonic acid', 'Sesquiterpenes', 'Acrylic acid', 'Oxylenol', 'Pbenzoquinone',
    'CO', 'O3', 'HNO3', 'NO2', 'NO', 'NOx', 'NO3', 
    'NONO2Ratio', 'PAN', 'PPN', 'HONO', 'MPN',
] 

# haven't figured out how to add MO2 and MCO3 into the figure... TBD work.

# 2) data directory and other settings, need to be customized
Flight_IDs = ['RF03', 'RF07', 'RF09', 'P-3B', 'FN19']
Flight_IDs = [ 'P-3B', 'FN19']
#Flight_IDs = [ 'RF07']
#Flight_IDs = ['RF03', 'RF07', 'RF09']
Flight_IDs = ['FN19']

#Flight_IDs = ['RF01', 'RF02', 'RF03', 'RF04', 'RF05', 
#              'RF06', 'RF07', 'RF09', 'RF10',
#              'RF11', 'RF13']

# 4) setting for smk or bkg
get_smk, get_smk_dil, get_smk_conorm = True, False, False
#get_smk, get_smk_dil, get_smk_conorm = False, True, False
#get_smk, get_smk_dil, get_smk_conorm = False, False, True


# chossing the mechanism you want to run 
# current choices include GC_base, GC_basenofixed, GC_modified, GC_modified_v2, MCM_base_FUR, MCM_base_noFUR, MCM_base_modified, MCM_GCvocs, and MCM_GCvocs_modified
#Running_mech = ['GC_base',  'MCM_base_FUR', 'MCM_base_noFUR', 'MCM_GCvocs'] 
#Running_mech = ['GC_base',  'MCM_base_FUR', 'MCM_base_noFUR', 'MCM_GCvocs'] 
Running_mech = ['GC_base', 'GC_modified', 'MCM_base_FUR', 'MCM_GCvocs'] 
Running_mech = ['GC_base', 'GC_modified', 'MCM_base_FUR', 'MCM_GCvocs', 'MCM_GCvocs_modified'] 

# 3) file directories
for Flight_ID in Flight_IDs:
    print(f'Flight ID: This is {Flight_ID}')
    if 'RF03' or 'RF07' or 'RF09' in Flight_ID:
        file_dir = '/glade/u/home/lixujin/matlab/F0AM-4.2.1/Setups/Examples/Lixu/WE-CAN/output_data/WECAN' 
    if 'P-3B' in Flight_ID:
        file_dir = '/glade/u/home/lixujin/matlab/F0AM-4.2.1/Setups/Examples/Lixu/P-3B/output_data/FIREX_LGPlume'
    if 'FN' in Flight_ID:
        file_dir = '/glade/u/home/lixujin/matlab/F0AM-4.2.1/Setups/Examples/Lixu/FIREX-AQ/output_data/FIREXAQ'
    
    if not obs_only:
        # 3.1) FOAM files for the mixing ratio output
        GC_base             = file_dir + '_extract_GCv13_base' + Flight_ID.replace('-', '') + '.mat'
        GC_MCMvocs          = file_dir + '_extract_GCv13_MCM_missing' + Flight_ID.replace('-', '') + '.mat'
        GC_basenofixed      = file_dir + '_extract_GCv13_base_notfixed' + Flight_ID.replace('-', '') + '.mat'
        GC_modified         = file_dir + '_extract_GCv13_JINF' + Flight_ID.replace('-', '') + '.mat'
        GC_modified_v2      = file_dir + '_extract_GCv13_JINF_HCOOH' + Flight_ID.replace('-', '') + '.mat'

        MCM_base_noFUR      = file_dir + '_extract_MCMv331_noFUR' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_FUR        = file_dir + '_extract_MCMv331_base' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_extension  = file_dir + '_extract_MCMv331_extension' + Flight_ID.replace('-', '') + '.mat'
        MCM_GCvocs          = file_dir + '_extract_MCMv331_GCvocs' + Flight_ID.replace('-', '') + '.mat'
        MCM_GCvocs_modified = file_dir + '_extract_MCMv331_GCvocsF' + Flight_ID.replace('-', '') + '.mat'

        # 3.2) FOAM files for the dilution corrected mixing ratio
        GC_base_dil             = file_dir + '_dilution_corrected_extract_' + 'GCv13_base' + Flight_ID.replace('-', '') + '.mat'
        GC_MCMvocs_dil          = file_dir + '_dilution_corrected_extract_' + 'GCv13_MCM_missing' + Flight_ID.replace('-', '') + '.mat'
        GC_basenofixed_dil      = file_dir + '_dilution_corrected_extract_' + 'GCv13_base_notfixed' + Flight_ID.replace('-', '') + '.mat'
        GC_modified_dil         = file_dir + '_dilution_corrected_extract_' + 'GCv13_JINF' + Flight_ID.replace('-', '') + '.mat'
        GC_modified_v2_dil      = file_dir + '_dilution_corrected_extract_' + 'GCv13_JINF_HCOOH' + Flight_ID.replace('-', '') + '.mat'

        MCM_base_noFUR_dil      = file_dir + '_dilution_corrected_extract_' + 'MCMv331_noFUR' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_FUR_dil        = file_dir + '_dilution_corrected_extract_' + 'MCMv331_base' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_extension_dil  = file_dir + '_dilution_corrected_extract_' + 'MCMv331_extension' + Flight_ID.replace('-', '') + '.mat'
        MCM_GCvocs_dil          = file_dir + '_dilution_corrected_extract_' + 'MCMv331_GCvocs' + Flight_ID.replace('-', '') + '.mat'
        MCM_GCvocs_modified_dil = file_dir + '_dilution_corrected_extract_' + 'MCMv331_GCvocsF' + Flight_ID.replace('-', '') + '.mat'

        # 3.3) F0AM files for CO normalized values
        GC_base_conorm             = file_dir + '_co_normalized_corrected_extract_' + 'GCv13_base' + Flight_ID.replace('-', '') + '.mat'
        GC_MCMvocs_conorm          = file_dir + '_co_normalized_corrected_extract_' + 'GCv13_MCM_missing' + Flight_ID.replace('-', '') + '.mat'
        GC_basenofixed_conorm      = file_dir + '_co_normalized_corrected_extract_' + 'GCv13_base_notfixed' + Flight_ID.replace('-', '') + '.mat'
        GC_modified_conorm         = file_dir + '_co_normalized_corrected_extract_' + 'GCv13_JINF' + Flight_ID.replace('-', '') + '.mat'
        GC_modified_v2_conorm      = file_dir + '_co_normalized_corrected_extract_' + 'GCv13_JINF_HCOOH' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_noFUR_conorm      = file_dir + '_co_normalized_corrected_extract_' + 'MCMv331_noFUR' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_FUR_conorm        = file_dir + '_co_normalized_corrected_extract_' + 'MCMv331_base' + Flight_ID.replace('-', '') + '.mat'
        MCM_base_extension_conorm  = file_dir + '_co_normalized_corrected_extract_' + 'MCMv331_extension' + Flight_ID.replace('-', '') + '.mat'
        MCM_GCvocs_conorm          = file_dir + '_co_normalized_corrected_extract_' + 'MCMv331_GCvocs' + Flight_ID.replace('-', '') + '.mat'
        MCM_GCvocs_modified_conorm = file_dir + '_co_normalized_corrected_extract_' + 'MCMv331_GCvocsF' + Flight_ID.replace('-', '') + '.mat'

    # 5) choose the plotting type
    tsplot = True
    scplot = False # there is a bug needs to be fixed.

    # 6) setting start_ind for time indexing
    start_ind = 1 if (Flight_ID == 'RF03') else 0
    
    
# =====================================================================================================
# load observed bkg data
# different flights have different ways to read in the observation data due to the format inconsistence 
# for example, WE-CAN data is loaded in csv data
# FIREX-AQ and DISCOVERY-AQ data is read in .mat as they are processed from Matt
# If we want to further add more observations, corresponding codes need to be added (future suggestion).
# =====================================================================================================
    # 7) Getting background values for observations.
    # 7.1) WE-CAN
    if 'RF' in Flight_ID:
        df_obs_bkg = get_obs_df(Flight_ID, spec_total, get_smk=False, get_bkg=True)
        # data processing
        test_min_bkg = True
        if test_min_bkg:
            for col in df_obs_bkg.columns:
                df_obs_bkg[col] = [min(df_obs_bkg[col]) for _ in range(len(df_obs_bkg.index))]
    # 7.2) P-3B
    if 'P-3B' in Flight_ID:
        data_bkg = [95, 0.031, 0.112, 35.4, 1897, 0.084, 1.345, 0.389, 0.008, 0.073, 0.219, 0.183*0.5, 0.183*0.5, 1.937, 1.491] # This is what Matt used.
        data_name = ['CO', 'NO', 'NO2', 'O3' , 'CH4', 'Propene', 'Formaldehyde', 'Acetaldehyde', 'Benzene', 'Acetic acid', 'Isoprene', 'Methacrolein', 'MVK', 'Methanol', 'Acetone']
        df_obs_bkg = pd.DataFrame(data=data_bkg, index=data_name).transpose()

        # make up some values based on the current knowledge of PTR
        #df_obs_bkg['Ethene'] = df_obs_bkg['C3H6']
        df_obs_bkg['Methacrolein']          = df_obs_bkg['MVK'] * 0.5
        df_obs_bkg['MVK']   = df_obs_bkg['MVK'] * 0.5
        #df_obs_bkg['Lumped C>=3 aldehydes'] = df_obs_bkg['Acetone'] * 0.22    
        #df_obs_bkg['Acetone']               = df_obs_bkg['Acetone'] * 0.78
        df_obs_bkg['Lumped C>=3 aldehydes'] = 0 
        df_obs_bkg['Acetone']               = 0
        df_obs_bkg['Isoprene']              = 0 # try to avoid negative NEMR
        df_obs_bkg['NOx'] = df_obs_bkg['NO'] + df_obs_bkg['NO2']
        df_obs_bkg['Furans']                = 0 
        df_obs_bkg['Furfurals']             = 0 


        # drop those unnecessary primary VOCs
        #df_obs_bkg = df_obs_bkg.drop(['CH4', 'C3H6', 'PRPE', 'Benzene', 'Methanol'], axis=1)
        #df_obs_bkg = df_obs_bkg.drop(['CH4', 'Benzene', 'Methanol'], axis=1)

        # fill the missing background values
        for col in spec_total:
            if col not in df_obs_bkg.columns: df_obs_bkg[col] = 0.

    # 7.3) FIREX-AQ
    # This is preliminary settings! In later stage, we are planing to used the reduced size data prior using the python script
    # This is expected to largely increase the speed.
    if 'FN19' in Flight_ID: 
        init_time = 9.76
        file =  '/glade/u/home/lixujin/matlab/F0AM-4.2.1/Setups/Examples/Lixu/FIREX-AQ/output_data/FIREXAQ_dilution_correctedMCMv331_noFURFN19.mat' 
        df_obs_bkg = get_df_specs(file_name=file, spec_list = spec_total, 
                            GC_setting=False, MCM_setting=True, get_conc=False) 
        df_obs_bkg.index = df_obs_bkg.index + init_time

# =====================================================================================================
# load observation data
# different flights have different ways to read in the observation data due to the format inconsistence 
# for example, WE-CAN data is loaded in csv data
# FIREX-AQ and DISCOVERY-AQ data is read in .mat as they are processed from Matt
# If we want to further add more observations, corresponding codes need to be added (future suggestion).
# =====================================================================================================
    # WE-CAN flights
    if 'RF' in Flight_ID:
        # observations
        df_obs_smk      = get_obs_df(Flight_ID, spec_total, get_smk=True, get_bkg = False)
        # calcualte the dilution/conorm corrected mixing ratio
        delta_CO = df_obs_smk['CO'] - df_obs_bkg['CO']
        fCO = delta_CO.iloc[start_ind] / delta_CO
        df_obs_smk_dil_corrected = pd.DataFrame().reindex_like(df_obs_smk)
        df_obs_smk_conorm_corrected = pd.DataFrame().reindex_like(df_obs_smk)
        for col in df_obs_smk.columns:
            delta_X = df_obs_smk[col] - df_obs_bkg[col]
            df_obs_smk_dil_corrected[col] = delta_X*fCO
            df_obs_smk_conorm_corrected[col] = df_obs_smk[col]/df_obs_smk['CO']
##################################################################################
# Read observation data including the background and smoke concentrationspec_total
##################################################################################
    # P-3B understory fires: this might be slighly more complex than reading csv files
    if 'P-3B' in Flight_ID:
        F0AM_model_file = '/glade/u/home/lixujin/matlab/F0AM-4.2.1/Setups/Examples/Lixu/P-3B/LagrangianPlumeData_Extrapolated.mat'
        mod_struc = scipy.io.loadmat(F0AM_model_file)
        # ==============
        # Variable names
        # ==============
        var_names = []
        # varaible in the data
        dtype_names = [('P', 'O'), ('T', 'O'), ('TIME', 'O'), ('RH', 'O'), ('SZA', 'O'), ('CH4', 'O'), ('NO', 'O'), ('NO2', 'O'), ('O3', 'O'), ('CO2', 'O'), ('CO', 'O'), ('HCHO', 'O'), ('HONO', 'O'), ('CH3CHO', 'O'), ('C3H6', 'O'), ('BENZENE', 'O'), ('FURAN', 'O'), ('C5H8', 'O'), ('CH3COCH3', 'O'), ('HCOOH', 'O'), ('CH3CO2H', 'O'), ('BIACET', 'O'), ('MVK', 'O'), ('MGLYOX', 'O'), ('FURFURAL', 'O'), ('ACETOL', 'O'), ('JNO2', 'O'), ('CH3OH', 'O'), ('DIL1', 'O'), ('jcorr', 'O'), ('Sol', 'O'), ('MEFURAN', 'O'), ('MALANHY', 'O'), ('C4H4O3', 'O'), ('FURANONE', 'O'), ('PHENOL', 'O'), ('CATECHOL', 'O'), ('STYRENE', 'O'), ('CRESOL', 'O'), ('TIME_EXT', 'O'), ('CO_EXT', 'O'), ('H2O', 'O')]
        for dname, tmp in dtype_names:
            var_names.append(dname)
        # remove the inconsistent data
        var_names.remove(var_names[30])

        # ==============
        # Variable data
        # ==============
        # retrieve the irregular information first
        loc_date = mod_struc['DAQ'].item()[30]
        lat, lon, alt = loc_date.item()[0][0][0], loc_date.item()[1][0][0], loc_date.item()[2][0][0]
        startTime, updateTime =  loc_date.item()[3][0][0], loc_date.item()[4][0][0]
        # other data
        var_data = []
        ct = 0
        for ele in mod_struc['DAQ'].item():
            if ct != 30: 
                ele = [item for sublist in ele for item in sublist]
                var_data.append(ele)
            # increment the pointer
            ct += 1

        # ================================================
        # Prepare met data
        # Only for model input but not for data analysis 
        # ================================================
        # Met data: get ind where len of element is 65
        ind_65dp = [len(num_dp) == 65 for num_dp in var_data]
        met_data  = np.array(var_data)[ind_65dp] 
        met_names = np.array(var_names)[ind_65dp] 
        df_met = pd.DataFrame(list(met_data), index = met_names).transpose()
        met_time = df_met['TIME_EXT']
        df_met.index = met_time/60 # sec to min
        df_met = df_met.drop('TIME_EXT', axis = 1)

        # smk data: get ind where len of element is 15
        ind_15dp = [len(num_dp) == 15 for num_dp in var_data]
        smk_data  = np.array(var_data)[ind_15dp] 
        smk_names = np.array(var_names)[ind_15dp] 
        df_obs_smk = pd.DataFrame(list(smk_data), index = smk_names).transpose()
        smk_time = df_obs_smk['TIME']
        df_obs_smk.index = smk_time/60 # sec to min

        # rename the column names
        df_obs_smk = df_obs_smk.rename(columns={"HCHO":"Formaldehyde", "CH3CHO": "Acetaldehyde", "BENZENE":"Benzene", 
                                                "FURAN":"Furan", "C5H8": "Isoprene", 
                                                "HCOOH": "Formic acid", "CH3CO2H": "Acetic acid", "BIACET": "2,3-butanedione",
                                                "MGLYOX": "Methylglyoxal", "FURFURAL": "Furfural", "ACETOL": "Hydroxyacetone", 
                                                "CH3OH": "Methanol", "MEFURAN": "Methylfuran", "MALANHY": "Maleic anhydride", 
                                                "C4H4O3": "5-hydroxy-2(5H)-furanone", "FURANONE": "Furanone", "PHENOL": "Phenol", 
                                                "CATECHOL": "Catechol", "STYRENE": "Styrene", "CRESOL": "Cresol"})

        # Using the ratio to scale up ethene, lumped aldehydes, MACR, MVK, acetone, dimethylfuran, methylfurfural, Phenol, Catechol, Cresol
        df_obs_smk['Ethene']                = df_obs_smk['C3H6']
        df_obs_smk['Lumped C>=3 aldehydes'] = df_obs_smk['CH3COCH3'] * 0.22
        df_obs_smk['Methacrolein']          = df_obs_smk['MVK'] * 0.5
        df_obs_smk['MVK']   = df_obs_smk['MVK'] * 0.5
        df_obs_smk['Acetone']               = df_obs_smk['CH3COCH3'] * 0.78

        #df_obs_smk['Dimethylfuran']         = df_obs_smk['Methylfuran'] * 0.5
        df_obs_smk['Methylfurfural']        = df_obs_smk['Catechol'] * 0.5
        df_obs_smk['Phenol']                = df_obs_smk['Phenol'] * 1.05
        df_obs_smk['Catechol']              = df_obs_smk['Catechol'] * 0.5
        df_obs_smk['Cresol']                = df_obs_smk['Cresol'] * 1.5
        df_obs_smk['NOx'] = df_obs_smk['NO'] + df_obs_smk['NO2']

        # drop some columns that are no needed.
        df_obs_smk = df_obs_smk.drop(['TIME', 'DIL1'], axis = 1)
        df_obs_smk = df_obs_smk.drop(['CH4', 'CO2', 'C3H6', 'Benzene', 'Methanol', 'Ethene', 'CH3COCH3',
                                      '5-hydroxy-2(5H)-furanone'], axis = 1)

        # ============================================================
        # calcualte dilution normalized mixing ratio for observations
        # ============================================================
        delta_CO = df_obs_smk['CO'] - df_obs_bkg['CO'].values
        fCO = delta_CO.iloc[0] / delta_CO
        df_obs_smk_dil_corrected = pd.DataFrame().reindex_like(df_obs_smk)
        df_obs_smk_conorm_corrected = pd.DataFrame().reindex_like(df_obs_smk)
        for col in df_obs_smk.columns:
            b = df_obs_bkg[col].values[0] if col in df_obs_bkg.columns else 0
            delta_X = df_obs_smk[col] - b
            df_obs_smk_dil_corrected[col] = delta_X*fCO  
            df_obs_smk_conorm_corrected[col] = df_obs_smk[col]/df_obs_smk['CO'] 

    # FIREX-AQ fires: this might be slighly more complex than reading csv files
    if 'FN19' in Flight_ID:
        # reading in chem input data
        F0AM_model_file_chem =  '/glade/u/home/lixujin/matlab/F0AM-4.2.1/Setups/Examples/Lixu/FIREX-AQ/FN19_Blackwater_20190830/ChemInput_avg_FN19Blackwater_20200513.mat' 
        mod_struc_chem = scipy.io.loadmat(F0AM_model_file_chem)
        data_dummy = []
        names = []
        for key in list(mod_struc_chem.keys()):
            if len(mod_struc_chem[key])==4: 
                data_dummy.append(mod_struc_chem[key])
                names.append(key)

        # mapping names in to the standard form   
        names_FIREXAQ2std = {'AceticAcid': 'Acetic acid',
                             'Acrylicacid': 'Acrylic acid',
                             'BenzeneDiol': 'Benzene Diol',
                             'FormicAcid': 'Formic acid', 
                             'MaleicAnhydride': 'Maleic anhydride',
                             'acetylene': 'Acetylene',
                             'butenes': 'Butenes',
                             'ethene': 'C2H4',
                             'hydroxyacetone': 'Hydroxyacetone',
                             'methylacetate': 'Methylacetate',
                             'MethylFuran': 'Methylfuran',
                             'MethylFurfural': 'Methylfurfural'}
        names = [names_FIREXAQ2std.get(key, key) for key in names]
        df_obs_smk = pd.DataFrame(np.hstack(data_dummy), columns = names)

        # set up index for the dataframe
        time = df_obs_smk['TimeDownwind_s']
        df_obs_smk = df_obs_smk.drop('TimeDownwind_s', axis = 1)
        df_obs_smk.index = time/60 # convert second to min
        # Correct the intial time of FN19
        df_obs_smk.index =  df_obs_smk.index + 9.67

        
        # adding values for lumping compounds
        df_obs_smk.loc[:, 'NOx'] = df_obs_smk.loc[:, 'NO'] +  df_obs_smk.loc[:, 'NO2']    
        df_obs_smk.loc[:, 'Lumped C>=3 alkenes'] = df_obs_smk.loc[:, 'Propene'] +  df_obs_smk.loc[:, 'Butenes'] +  df_obs_smk.loc[:, 'Pentene_methylbutene']

        df_obs_smk.loc[:, 'Furans'] = df_obs_smk.loc[:, 'Furan'] +  df_obs_smk.loc[:, 'Methylfuran'] +  df_obs_smk.loc[:, 'Dimethylfuran']   
        df_obs_smk.loc[:, 'Furfurals'] = df_obs_smk.loc[:, 'Furfural'] +  df_obs_smk.loc[:, 'Methylfurfural'] 
        # reduced observation to only species we are interested in
        #df_obs_smk = df_obs_smk[spec_total]

        # ============================================================
        # calcualte dilution normalized mixing ratio for observations
        # ============================================================
        delta_CO = df_obs_smk['CO'] - df_obs_bkg['CO'].values[0]
        fCO = delta_CO.iloc[0] / delta_CO
        df_obs_smk_dil_corrected = pd.DataFrame().reindex_like(df_obs_smk)
        df_obs_smk_conorm_corrected = pd.DataFrame().reindex_like(df_obs_smk)
        for col in df_obs_smk.columns:
            b = df_obs_bkg[col].values[0] if col in df_obs_bkg.columns else 0
            delta_X = df_obs_smk[col] - b
            df_obs_smk_dil_corrected[col] = delta_X*fCO 
            df_obs_smk_conorm_corrected[col] = df_obs_smk[col]/df_obs_smk['CO']

########################
# Read model simulations
########################
    # ===========================================================================================
    # 2) Dilution corrected mixing ratio
    # calculate normalized excess mixing ratios: delta_X_dil = delta_X * delta_CO_source/delta_CO
    # ===========================================================================================
    if get_smk:
        # intialized the data
        df_obs_smk_conc = df_obs_smk
        init_time = df_obs_smk_conc.index[start_ind]
        standard_index = np.array(df_obs_smk_conc.index[start_ind:-1], dtype='float64')
        if not obs_only:
            # GEOS-Chem base
            if 'GC_base' in Running_mech:
                print('This is for GEOS-Chem (base)')
                df_gc_base_smk_conc = F0AM_cleaner_reduced(file_name=GC_base, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_base_smk_conc_interp = interp(df_gc_base_smk_conc, standard_index)

            # GEOS-Chem, MCM VOCs... We may need to add those different names up!!! Check it later
            if 'GC_MCMvocs' in Running_mech:
                print('This is for GEOS-Chem (MCM VOCs)')
                df_gc_MCMvocs_smk_conc = F0AM_cleaner_reduced(file_name=GC_MCMvocs, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_MCMvocs_smk_conc_interp = interp(df_gc_MCMvocs_smk_conc, standard_index)

            # GEOS-Chem nofixed
            if 'GC_basenofixed' in Running_mech:
                print('This is for GEOS-Chem (nofixed)')
                df_gc_basenofixed_smk_conc = F0AM_cleaner_reduced(file_name=GC_basenofixed, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_basenofixed_smk_conc_interp = interp(df_gc_basenofixed_smk_conc, standard_index)
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:
                print('This is for GEOS-Chem (FUR)')
                df_gc_modi_smk_conc        = F0AM_cleaner_reduced(file_name=GC_modified, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_modi_smk_conc_interp = interp(df_gc_modi_smk_conc, standard_index)
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech:
                print('This is for GEOS-Chem (FUR + SW mech)')
                df_gc_modi_v2_smk_conc        = F0AM_cleaner_reduced(file_name=GC_modified_v2, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_modi_v2_smk_conc_interp = interp(df_gc_modi_v2_smk_conc, standard_index)    
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_base_noFUR' in Running_mech:  
                print('This is for MCM + noFUR')
                df_mcm_base_noFUR_smk_conc        = F0AM_cleaner_reduced(file_name=MCM_base_noFUR, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_noFUR_smk_conc_interp = interp(df_mcm_base_noFUR_smk_conc, standard_index)
            # MCM base
            if 'MCM_base_FUR' in Running_mech:
                print('This is for MCM + FUR')
                df_mcm_base_FUR_smk_conc        = F0AM_cleaner_reduced(file_name=MCM_base_FUR, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_FUR_smk_conc_interp = interp(df_mcm_base_FUR_smk_conc, standard_index)

            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:
                print('This is for MCM + extended VOCs')
                df_mcm_base_extension_smk_conc        = F0AM_cleaner_reduced(file_name=MCM_base_extension, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_extension_smk_conc_interp = interp(df_mcm_base_extension_smk_conc, standard_index)

            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:    
                print('This is for MCM (GEOS-Chem VOCs)')
                df_mcm_GCvocs_smk_conc        = F0AM_cleaner_reduced(file_name=MCM_GCvocs, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_GCvocs_smk_conc_interp = interp(df_mcm_GCvocs_smk_conc, standard_index)
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech:    
                print('This is for MCM (GEOS-Chem VOCs + FUR)')
                df_mcm_GCvocs_modified_smk_conc        = F0AM_cleaner_reduced(file_name=MCM_GCvocs_modified, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_GCvocs_modified_smk_conc_interp = interp(df_mcm_GCvocs_modified_smk_conc, standard_index)

    if get_smk_dil:
        # intialized the data
        df_obs_smk_dil_conc = df_obs_smk_dil_corrected
        init_time = df_obs_smk_dil_conc.index[start_ind]
        standard_index = np.array(df_obs_smk_dil_conc.index[start_ind:-1], dtype='float64')
        if not obs_only:
            # GEOS-Chem base
            if 'GC_base' in Running_mech:
                print('This is for GEOS-Chem (base)')
                df_gc_base_smk_dil_conc = F0AM_cleaner_reduced(file_name=GC_base_dil, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_base_smk_dil_conc_interp = interp(df_gc_base_smk_dil_conc, standard_index)
            # GEOS-Chem MCM VOCs
            if 'GC_MCMvocs' in Running_mech:
                print('This is for GEOS-Chem (MCM VOCs)')
                df_gc_MCMvocs_smk_dil_conc = F0AM_cleaner_reduced(file_name=GC_MCMvocs_dil, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_MCMvocs_smk_dil_conc_interp = interp(df_gc_MCMvocs_smk_dil_conc, standard_index)

            # GEOS-Chem nofixed
            if 'GC_basenofixed' in Running_mech:
                print('This is for GEOS-Chem (nofixed)')
                df_gc_basenofixed_smk_dil_conc = F0AM_cleaner_reduced(file_name=GC_basenofixed_dil, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_basenofixed_smk_dil_conc_interp = interp(df_gc_basenofixed_smk_dil_conc, standard_index)
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:
                print('This is for GEOS-Chem (FUR)')
                df_gc_modi_smk_dil_conc        = F0AM_cleaner_reduced(file_name=GC_modified_dil, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_modi_smk_dil_conc_interp = interp(df_gc_modi_smk_dil_conc, standard_index)
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech:
                print('This is for GEOS-Chem (FUR + SW mech)')
                df_gc_modi_v2_smk_dil_conc        = F0AM_cleaner_reduced(file_name=GC_modified_v2_dil, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_modi_v2_smk_dil_conc_interp = interp(df_gc_modi_v2_smk_dil_conc, standard_index)    
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_base_noFUR' in Running_mech:  
                print('This is for MCM + noFUR')
                df_mcm_base_noFUR_smk_dil_conc        = F0AM_cleaner_reduced(file_name=MCM_base_noFUR_dil, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_noFUR_smk_dil_conc_interp = interp(df_mcm_base_noFUR_smk_dil_conc, standard_index)
            # MCM base
            if 'MCM_base_FUR' in Running_mech:
                print('This is for MCM + FUR')
                df_mcm_base_FUR_smk_dil_conc        = F0AM_cleaner_reduced(file_name=MCM_base_FUR_dil, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_FUR_smk_dil_conc_interp = interp(df_mcm_base_FUR_smk_dil_conc, standard_index)

            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:
                print('This is for MCM + extended VOCs')
                df_mcm_base_extension_smk_dil_conc        = F0AM_cleaner_reduced(file_name=MCM_base_extension_dil, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_extension_smk_dil_conc_interp = interp(df_mcm_base_extension_smk_dil_conc, standard_index)

            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:    
                print('This is for MCM (GEOS-Chem VOCs)')
                df_mcm_GCvocs_smk_dil_conc        = F0AM_cleaner_reduced(file_name=MCM_GCvocs_dil, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_GCvocs_smk_dil_conc_interp = interp(df_mcm_GCvocs_smk_dil_conc, standard_index)
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech:    
                print('This is for MCM (GEOS-Chem VOCs + FUR)')
                df_mcm_GCvocs_modified_smk_dil_conc        = F0AM_cleaner_reduced(file_name=MCM_GCvocs_modified_dil, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_GCvocs_modified_smk_dil_conc_interp = interp(df_mcm_GCvocs_modified_smk_dil_conc, standard_index)

    if get_smk_conorm:
        # intialized the data
        df_obs_smk_conorm_conc = df_obs_smk_conorm_corrected
        init_time = df_obs_smk_conorm_conc.index[start_ind]
        standard_index = np.array(df_obs_smk_conorm_conc.index[start_ind:-1], dtype='float64')
        if not obs_only:
            # GEOS-Chem base
            if 'GC_base' in Running_mech:
                print('This is for GEOS-Chem (base)')
                df_gc_base_smk_conorm_conc = F0AM_cleaner_reduced(file_name=GC_base_conorm, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_base_smk_conorm_conc_interp = interp(df_gc_base_smk_conorm_conc, standard_index)
            # GEOS-Chem MCM VOCs
            if 'GC_MCMvocs' in Running_mech:
                print('This is for GEOS-Chem (MCM VOCs)')
                df_gc_MCMvocs_smk_conorm_conc = F0AM_cleaner_reduced(file_name=GC_MCMvocs_conorm, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_MCMvocs_smk_conorm_conc_interp = interp(df_gc_MCMvocs_smk_conorm_conc, standard_index)

            # GEOS-Chem nofixed
            if 'GC_basenofixed' in Running_mech:
                print('This is for GEOS-Chem (nofixed)')
                df_gc_basenofixed_smk_conorm_conc = F0AM_cleaner_reduced(file_name=GC_basenofixed_conorm, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_basenofixed_smk_conorm_conc_interp = interp(df_gc_basenofixed_smk_conorm_conc, standard_index)
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:
                print('This is for GEOS-Chem (FUR)')
                df_gc_modi_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=GC_modified_conorm, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_modi_smk_conorm_conc_interp = interp(df_gc_modi_smk_conorm_conc, standard_index)
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech:
                print('This is for GEOS-Chem (FUR + SW mech)')
                df_gc_modi_v2_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=GC_modified_v2_conorm, spec_list = spec_total, GC_setting=True, MCM_setting=False, init_time = init_time)
                df_gc_modi_v2_smk_conorm_conc_interp = interp(df_gc_modi_v2_smk_conorm_conc, standard_index)    
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_base_noFUR' in Running_mech:  
                print('This is for MCM + noFUR')
                df_mcm_base_noFUR_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=MCM_base_noFUR_conorm, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_noFUR_smk_conorm_conc_interp = interp(df_mcm_base_noFUR_smk_conorm_conc, standard_index)
            # MCM base
            if 'MCM_base_FUR' in Running_mech:
                print('This is for MCM + FUR')
                df_mcm_base_FUR_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=MCM_base_FUR_conorm, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_FUR_smk_conorm_conc_interp = interp(df_mcm_base_FUR_smk_conorm_conc, standard_index)

            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:
                print('This is for MCM + extended VOCs')
                df_mcm_base_extension_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=MCM_base_extension_conorm, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_base_extension_smk_conorm_conc_interp = interp(df_mcm_base_extension_smk_conorm_conc, standard_index)

            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:    
                print('This is for MCM (GEOS-Chem VOCs)')
                df_mcm_GCvocs_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=MCM_GCvocs_conorm, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_GCvocs_smk_conorm_conc_interp = interp(df_mcm_GCvocs_smk_conorm_conc, standard_index)
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech:    
                print('This is for MCM (GEOS-Chem VOCs + FUR)')
                df_mcm_GCvocs_modified_smk_conorm_conc        = F0AM_cleaner_reduced(file_name=MCM_GCvocs_modified_conorm, spec_list = spec_total, GC_setting=False, MCM_setting=True, init_time = init_time)
                df_mcm_GCvocs_modified_smk_conorm_conc_interp = interp(df_mcm_GCvocs_modified_smk_conorm_conc, standard_index)
    
    ###########################
    # set up data to be saved
    ###########################
    if get_smk:
        # Settign for plots
        subtitle = ' (Mixing ratio)'
        # Observation
        df_obs_conc = df_obs_smk
        if not obs_only:
            # GEOS-Chem base
            if 'GC_base' in Running_mech:
                df_gc_base_conc = df_gc_base_smk_conc
                df_gc_base_conc_interp = df_gc_base_smk_conc_interp
            # GEOS-Chem base
            if 'GC_MCMvocs' in Running_mech:
                print('This is for GEOS-Chem (MCM VOCs)')
                df_gc_MCMvocs_conc = df_gc_MCMvocs_smk_conc
                df_gc_MCMvocs_conc_interp = df_gc_MCMvocs_smk_conc_interp
            # GEOS-Chem basenofixed
            if 'GC_basenofixed' in Running_mech:
                df_gc_basenofixed_conc = df_gc_basenofixed_smk_conc
                df_gc_basenofixed_conc_interp = df_gc_basenofixed_smk_conc_interp
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:
                df_gc_modi_conc = df_gc_modi_smk_conc
                df_gc_modi_conc_interp = df_gc_modi_smk_conc_interp
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech:
                df_gc_modi_v2_conc = df_gc_modi_v2_smk_conc    
                df_gc_modi_v2_conc_interp = df_gc_modi_v2_smk_conc_interp
            # MCM mechanism with out FUR chemistry
            if 'MCM_base_noFUR' in Running_mech:
                df_mcm_base_noFUR_conc = df_mcm_base_noFUR_smk_conc
                df_mcm_base_noFUR_conc_interp = df_mcm_base_noFUR_smk_conc_interp
            # MCM base
            if 'MCM_base_FUR' in Running_mech:
                df_mcm_base_FUR_conc = df_mcm_base_FUR_smk_conc
                df_mcm_base_FUR_conc_interp = df_mcm_base_FUR_smk_conc_interp
            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:
                df_mcm_base_extension_conc = df_mcm_base_extension_smk_conc
                df_mcm_base_extension_conc_interp = df_mcm_base_extension_smk_conc_interp
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:
                df_mcm_GCvocs_conc = df_mcm_GCvocs_smk_conc
                df_mcm_GCvocs_conc_interp = df_mcm_GCvocs_smk_conc_interp
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech:
                df_mcm_GCvocs_modified_conc = df_mcm_GCvocs_modified_smk_conc
                df_mcm_GCvocs_modified_conc_interp = df_mcm_GCvocs_modified_smk_conc_interp
    if get_smk_dil:
        # Settign for plots
        subtitle = ' (Mixing ratio)'
        # Observation
        df_obs_conc = df_obs_smk_dil_conc
        if not obs_only:
            # GEOS-Chem base
            if 'GC_base' in Running_mech:
                df_gc_base_conc = df_gc_base_smk_dil_conc
                df_gc_base_conc_interp = df_gc_base_smk_dil_conc_interp
            # GEOS-Chem (MCM VOCs)
            if 'GC_MCMvocs' in Running_mech:
                df_gc_MCMvocs_conc = df_gc_MCMvocs_smk_dil_conc
                df_gc_MCMvocs_conc_interp = df_gc_MCMvocs_smk_dil_conc_interp
            # GEOS-Chem basenofixed
            if 'GC_basenofixed' in Running_mech:
                df_gc_basenofixed_conc = df_gc_basenofixed_smk_dil_conc
                df_gc_basenofixed_conc_interp = df_gc_basenofixed_smk_dil_conc_interp
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:
                df_gc_modi_conc = df_gc_modi_smk_dil_conc
                df_gc_modi_conc_interp = df_gc_modi_smk_dil_conc_interp
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech:
                df_gc_modi_v2_conc = df_gc_modi_v2_smk_dil_conc    
                df_gc_modi_v2_conc_interp = df_gc_modi_v2_smk_dil_conc_interp
            # MCM mechanism with out FUR chemistry
            if 'MCM_base_noFUR' in Running_mech:
                df_mcm_base_noFUR_conc = df_mcm_base_noFUR_smk_dil_conc
                df_mcm_base_noFUR_conc_interp = df_mcm_base_noFUR_smk_dil_conc_interp
            # MCM base
            if 'MCM_base_FUR' in Running_mech:
                df_mcm_base_FUR_conc = df_mcm_base_FUR_smk_dil_conc
                df_mcm_base_FUR_conc_interp = df_mcm_base_FUR_smk_dil_conc_interp
            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:
                df_mcm_base_extension_conc = df_mcm_base_extension_smk_dil_conc
                df_mcm_base_extension_conc_interp = df_mcm_base_extension_smk_dil_conc_interp
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:
                df_mcm_GCvocs_conc = df_mcm_GCvocs_smk_dil_conc
                df_mcm_GCvocs_conc_interp = df_mcm_GCvocs_smk_dil_conc_interp
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech:
                df_mcm_GCvocs_modified_conc = df_mcm_GCvocs_modified_smk_dil_conc
                df_mcm_GCvocs_modified_conc_interp = df_mcm_GCvocs_modified_smk_dil_conc_interp

    if get_smk_conorm:
        # Settign for plots
        subtitle = ' (Mixing ratio)'
        # Observation
        df_obs_conc = df_obs_smk_conorm_conc
        
        if not obs_only:
            # GEOS-Chem base
            if 'GC_base' in Running_mech:
                df_gc_base_conc = df_gc_base_smk_conorm_conc
                df_gc_base_conc_interp = df_gc_base_smk_conorm_conc_interp
            # GEOS-Chem (MCM VOCs)
            if 'GC_MCMvocs' in Running_mech:
                df_gc_MCMvocs_conc = df_gc_MCMvocs_smk_conorm_conc
                df_gc_MCMvocs_conc_interp = df_gc_MCMvocs_smk_conorm_conc_interp
            # GEOS-Chem basenofixed
            if 'GC_basenofixed' in Running_mech:
                df_gc_basenofixed_conc = df_gc_basenofixed_smk_conorm_conc
                df_gc_basenofixed_conc_interp = df_gc_basenofixed_smk_conorm_conc_interp
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:
                df_gc_modi_conc = df_gc_modi_smk_conorm_conc
                df_gc_modi_conc_interp = df_gc_modi_smk_conorm_conc_interp
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech:
                df_gc_modi_v2_conc = df_gc_modi_v2_smk_conorm_conc    
                df_gc_modi_v2_conc_interp = df_gc_modi_v2_smk_conorm_conc_interp
            # MCM mechanism with out FUR chemistry
            if 'MCM_base_noFUR' in Running_mech:
                df_mcm_base_noFUR_conc = df_mcm_base_noFUR_smk_conorm_conc
                df_mcm_base_noFUR_conc_interp = df_mcm_base_noFUR_smk_conorm_conc_interp
            # MCM base
            if 'MCM_base_FUR' in Running_mech:
                df_mcm_base_FUR_conc = df_mcm_base_FUR_smk_conorm_conc
                df_mcm_base_FUR_conc_interp = df_mcm_base_FUR_smk_conorm_conc_interp
            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:
                df_mcm_base_extension_conc = df_mcm_base_extension_smk_conorm_conc
                df_mcm_base_extension_conc_interp = df_mcm_base_extension_smk_conorm_conc_interp
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:
                df_mcm_GCvocs_conc = df_mcm_GCvocs_smk_conorm_conc
                df_mcm_GCvocs_conc_interp = df_mcm_GCvocs_smk_conorm_conc_interp
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech:
                df_mcm_GCvocs_modified_conc = df_mcm_GCvocs_modified_smk_conorm_conc
                df_mcm_GCvocs_modified_conc_interp = df_mcm_GCvocs_modified_smk_conorm_conc_interpr

    ###########################
    # Save out the data
    ###########################
    # make model results have the same number of data
    if not obs_only:
        if 'GC_base' in Running_mech:
            compounds_interest = df_gc_base_conc.columns
            standard_index = df_gc_base_conc.index # the one with least time
            # GEOS-Chem base
            if 'GC_base' in Running_mech:        df_gc_base_conc_std    = interp(df_gc_base_conc[compounds_interest], standard_index)
            # GEOS-Chem MCM VOCs
            if 'GC_MCMvocs' in Running_mech:        df_gc_MCMvocs_conc_std    = interp(df_gc_MCMvocs_conc[compounds_interest], standard_index)
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:    df_gc_modi_conc_std    = interp(df_gc_modi_conc[compounds_interest], standard_index)
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech: df_gc_modi_v2_conc_std = interp(df_gc_modi_v2_conc[compounds_interest], standard_index)
            # MCM base
            if 'MCM_base_FUR' in Running_mech:   df_mcm_base_FUR_conc_std    = interp(df_mcm_base_FUR_conc[compounds_interest], standard_index)
            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:   df_mcm_base_extension_conc_std    = interp(df_mcm_base_extension_conc[compounds_interest], standard_index)
            # MCM mechanism with out FUR chemistry
            if 'MCM_base_noFUR' in Running_mech: df_mcm_base_noFUR_conc_std  = interp(df_mcm_base_noFUR_conc[compounds_interest], standard_index)
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:     df_mcm_GCvocs_conc_std      = interp(df_mcm_GCvocs_conc[compounds_interest], standard_index)
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech: df_mcm_GCvocs_modified_conc_std   = interp(df_mcm_GCvocs_modified_conc[compounds_interest], standard_index)        
###########################
# Create the directory
###########################
    if 'RF' in Flight_ID:
        parent_dir = '/glade/work/lixujin/PYTHON/SciProj/Box_modeling_analysis/F0AM_analysis_TS/WE-CAN/Dataprocess/analysis_bycompound'
    if 'P-3B' in Flight_ID:
        parent_dir = '/glade/work/lixujin/PYTHON/SciProj/Box_modeling_analysis/F0AM_analysis_TS/P-3B/Dataprocess/analysis_bycompound'
    if 'FN' in Flight_ID:
        parent_dir = '/glade/work/lixujin/PYTHON/SciProj/Box_modeling_analysis/F0AM_analysis_TS/FIREX-AQ/Dataprocess/analysis_bycompound'
    compounds_interest = spec_total #+ ['MO2', 'MCO3', 'HO2', 'OH', 'HOx', 'TRO2', 'TPNs', 'TANs', 'TotherNOy', 'TVOCR','TNOy', 'TNOz']

    for compound in compounds_interest:
        path = os.path.join(parent_dir, compound).replace(' ', '_')
        try:
            os.mkdir(path)
            print("Directory '%s' created" %compound)
        except:
            print("Directory '%s' already exist" %compound)

        # make the data frame for modeled maleic anhydride
        if compound in df_obs_conc.columns: data_obs = {'Observation': df_obs_conc[compound]}
        if compound not in df_obs_conc.columns: data_obs = {'Observation': [0]*len(df_obs_conc)}
        if compound == 'NONO2Ratio': data_obs = {'Observation': df_obs_conc['NO']/df_obs_conc['NO2']}
        # Mute or make editions to serveral compounds
        # Applicable to all
        #if compound == 'Glyoxal': data_obs = {'Observation': df_obs_conc['CO']*0}
        if compound == 'Glyoxal':
            if Flight_ID == 'RF03': data_obs = {'Observation': df_obs_conc[compound]*93} # scale it up based ERs
            if Flight_ID == 'RF07': data_obs = {'Observation': df_obs_conc[compound]*371} # scale it up based ERs
            if Flight_ID == 'RF09': data_obs = {'Observation': df_obs_conc[compound]*88} # scale it up based ERs
            if Flight_ID == 'FN19': data_obs = {'Observation': df_obs_conc[compound]*2} # scale it up based ERs
        if 'RF' in Flight_ID and compound == 'Isoprene': data_obs = {'Observation': df_obs_conc['CO']*0}
        if 'RF' in Flight_ID and compound == 'Ethanol': data_obs = {'Observation': df_obs_conc['CO']*0}
        if 'RF' in Flight_ID and compound == 'DMS': data_obs = {'Observation': df_obs_conc['CO']*0}
        if 'RF' in Flight_ID and compound == 'Trimethyl123benzene': data_obs = {'Observation': df_obs_conc['CO']*0}
        if 'RF' in Flight_ID and compound == 'Styrene': data_obs = {'Observation': df_obs_conc['CO']*0}
        if not obs_only:
            data_mod = {}
            # GEOS-Chem base
            if 'GC_base' in Running_mech:        data_mod = {**data_mod, **{'GEOS-Chem (base)': df_gc_base_conc_std[compound]}}
            # GEOS-Chem MCM VOCs
            if 'GC_MCMvocs' in Running_mech:     data_mod = {**data_mod, **{'GEOS-Chem + MCM VOCs': df_gc_MCMvocs_conc_std[compound]}}
            # GEOS-Chem base experiment 1
            if 'GC_modified' in Running_mech:    data_mod = {**data_mod, **{'GEOS-Chem + FUR': df_gc_modi_conc_std[compound]}}
            # GEOS-Chem base experiment 2
            if 'GC_modified_v2' in Running_mech: data_mod = {**data_mod, **{'GEOS-Chem + FUR + others': df_gc_modi_v2_conc_std[compound]}}
            # MCM base
            if 'MCM_base_FUR' in Running_mech:   data_mod = {**data_mod, **{'MCM + FUR': df_mcm_base_FUR_conc_std[compound]}}
            # MCM base + extended VOCs
            if 'MCM_base_extension' in Running_mech:   data_mod = {**data_mod, **{'MCM + extended VOCs': df_mcm_base_extension_conc_std[compound]}}
            # MCM mechanism with out FUR chemistry
            if 'MCM_base_noFUR' in Running_mech: data_mod = {**data_mod, **{'MCM + noFUR (base)': df_mcm_base_noFUR_conc_std[compound]}}
            # MCM mechanism but only with GEOS-Chem VOCs
            if 'MCM_GCvocs' in Running_mech:     data_mod = {**data_mod, **{'MCM + GEOS-Chem VOCs': df_mcm_GCvocs_conc_std[compound]}}
            # MCM mechanism but only with GEOS-Chem VOCs and another experiment test
            if 'MCM_GCvocs_modified' in Running_mech: data_mod = {**data_mod, **{'MCM + GEOS-Chem VOCs + FUR': df_mcm_GCvocs_modified_conc_std[compound]}}

        # creates pandas DataFrame and save out the data
        df_obs = pd.DataFrame(data_obs, index=df_obs_conc.index)
        if not obs_only: df_mod = pd.DataFrame(data_mod, index=standard_index)
        # standardrize the observation index
        df_obs.index = pd.Series(df_obs.index).round(2)
        # save out the data
        if get_smk:
            df_obs.to_csv(path + '/' + Flight_ID + '_obs_smk.csv') 
            if not obs_only: df_mod.to_csv(path + '/' + Flight_ID + '_mod_smk.csv') 
        if get_smk_dil:
            df_obs.to_csv(path + '/' + Flight_ID + '_obs_smk_dil.csv') 
            if not obs_only: df_mod.to_csv(path + '/' + Flight_ID + '_mod_smk_dil.csv')
        if get_smk_conorm:
            df_obs.to_csv(path + '/' + Flight_ID + '_obs_smk_conorm.csv') 
            if not obs_only: df_mod.to_csv(path + '/' + Flight_ID + '_mod_smk_conorm.csv')
```
